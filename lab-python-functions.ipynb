```python
#1.
def initialize_inventory(products):
    """
    TO DO 
    Define a function named `initialize_inventory` 
    that takes `products` as a parameter. Inside the function, implement 
    the code for initializing the inventory dictionary using a loop and user input.
    """
    inventory = {}
    for product in products : 
        quantity_products = int(input(f"Enter the quantity of {product} : "))
        inventory[product] = quantity_products
    return inventory

    
#2
def get_customer_orders():
    """
    TO DO 
    Define a function named `get_customer_orders` that takes no parameters. 
    Inside the function, implement the code for prompting the user to enter 
    the product names using a loop. 
    The function should return the `customer_orders` set.
    """
    customer_orders = set()
    
    while True:
       
        order = input("Enter the name of the product you want in your order")
        safety_order = order.lower()

        if safety_order :
            customer_orders.add(safety_order)
            confirmation = input("Do want to order more ? yes or no ?").strip().lower()

            if confirmation == "no": 
                break
        else:
            print("product not available")
            break

    return customer_orders

#3. 
def update_inventory(customer_orders,inventory):
    """
    TO DO
    Define a function named `update_inventory` that takes `customer_orders` 
    and `inventory` as parameters. Inside the function, 
    implement the code for updating the inventory dictionary based on the customer orders.
    """
    for i in customer_orders:
        if inventory[i] > 0 and inventory[i] > 0:
            inventory[i] -= 1

    return inventory


#4. 
def calculate_order_statistics(customer_orders,products):
    """
    TO DO
    Define a function named `calculate_order_statistics` that takes `customer_orders` 
    and `products` as parameters. Inside the function, 
    implement the code for calculating the order statistics 
    (total products ordered, and percentage of unique products ordered). 
    The function should return these values.
    """
    total_order = len(customer_orders)

    percentage_order = (total_order/len(products))*100

    print(f"total order = {total_order},percentage = {percentage_order}")

    return total_order, percentage_order


#5.
def print_order_statistics(order_statistics):
    """  
    TO DO
    Define a function named `print_order_statistics` 
    that takes `order_statistics` as a parameter. 
    Inside the function, implement the code for printing the order statistics.
    """
    print("order statistics:")
    print(f"Total products ordered : {order_statistics[0]}")
    print(f"Percentage of products ordered {order_statistics[1]}%")



#6.
def print_updated_inventory(inventory):
    """
    TO DO
    Define a function named `print_updated_inventory` 
    that takes `inventory` as a parameter. 
    Inside the function, implement the code for printing the updated inventory.
    """
    for i,y in inventory.items():
        print(f"product : {i}, quantity : {y}")

#7. 
def lauch_sequence ():
    """
    TO DO
    Call the functions in the appropriate sequence 
    to execute the program and manage customer orders.
    """
    products = ["t-shirt", "mug", "hat", "book", "keychain"]
    inventory = initialize_inventory(products)
    customer_orders = get_customer_orders()
    order_statistics = calculate_order_statistics(customer_orders, products)
    print_order_statistics(order_statistics)
    updated_inventory = update_inventory(customer_orders, inventory)
    print_updated_inventory(updated_inventory)

lauch_sequence()
```
